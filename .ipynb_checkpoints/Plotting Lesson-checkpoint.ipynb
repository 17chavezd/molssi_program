```python
import os
```

```python
import numpy
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
```

```python
print(headers)
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
sample = headers[1]
sample2 = headers[2]
sample3 = headers[3]
sample4 = headers[4]

plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
plt.plot(data[:,3], label=sample3)
plt.plot(data[:,4], label=sample4)
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend() 

plt.savefig('{four_samples}.png', dpi=300)
```

```python
num_columns = len(headers)
print(headers)
print(F'The number of columns is {num_columns}')
for col in range (1, num_columns):
    print(F'The column number is {col}.')
    plt.plot(data[:,col], label=headers[col])
    
plt.legend ()    
plt.xlabel('Simulation Frame') 
plt.ylabel('Distance (angstrom)')
plt.savefig('{All_Samples}', dpi=300)
```

```python
num_columns = len(headers)
print(headers)
print(F'The number of columns is {num_columns}')
for col in range (1, num_columns):
    plt.figure(col)
    plt.plot(data[:,col], label=headers[col])
    plt.legend ()    
    plt.xlabel('Simulation Frame') 
    plt.ylabel('Distance (angstrom)')
    #plt.savefig(F'{headers[col]}.png, dpi=300)
```
